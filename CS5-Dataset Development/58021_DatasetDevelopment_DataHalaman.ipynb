## Visualization of Dataset

The dataset below in the pie chart shows the sorted ages of the respondents with age count. The numbers outside the pie chart are the age of the respondents given from the survey while the percentage inside the pie chart is the number count of ages summarized using percentage.

```python
age_count = age_sort['Age'].value_counts()
# age_count.plot(xlabel="Age", ylabel="Count of Respondents", kind="bar", figsize=(15, 7))
age_count.plot(kind="pie", autopct='%1.1f%%', figsize=(10, 7))
plt.show()
```

The pie chart below shows the data of the license type count of the respondents. There are three choices of types of licenses given in the survey, it is then shown here that most of the respondents have non-professional driverâs licenses which 77.3% of them answered it and the followed by 23.3% for professional driverâs licenses and lastly, 3.3% only of the respondents use student driverâs permit.

```python
license_type_count.plot(kind="pie", autopct='%1.1f%%', figsize=(10, 7))
plt.legend()
```

The last dataset in pie chart shows the visualization of respondentsâ choice of vehicle classification. This counts the respondentâs answers and summarized them into percentages in order to plot them into a graph and visualize the data A huge part of the blue of the graph which consists of 80% of the respondentâs answers choosing class one has the majority of the answers chosen by the respondents and then followed by the orange part which has 16.7% which classified as class two, and only 3.3% for the green part which is class three.

```python
classification_count.plot(kind="pie", autopct='%1.1f%%', figsize=(10, 10))
plt.legend()
```

The dataset below shows the result within the following questions, âHow enough traffic signs, traffic lights, and traffic enforcers in your area?â. The result was shown in bar graph form in order to visualize the respondentâs answers. Here the percentage of âNot quite enough for the traffic signs has the majority of the respondentâs answers, the percentage of âEnoughâ for the traffic lights has the highest bar, and also the percentage of âEnoughâ for the traffic enforcers got the highest votes from the respondents.

```python
enough_signs_count = research_data_df[['How enough traffic signs in your area as: ']].value_counts()
enough_lights_count = research_data_df[['How enough traffic lights in your area as:']].value_counts()
enough_enforcer_count = research_data_df[['How enough traffic enforcers in your area as:']].value_counts()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
enough_signs_count.plot(kind='bar',ax=axes[0],figsize=(20,5))
enough_lights_count.plot(kind='bar',ax=axes[1],figsize=(20,5))
enough_enforcer_count.plot(kind='bar',ax=axes[2],figsize=(20,5))
```

Below shows the dataset in bar graph visualizations of the following questions âHow important are traffic signs, traffic lights, and traffic enforcers in your area?â. The bar graph represents the respondentâs answers and here, it shows that within the traffic signs, the majority of respondents have agreed that traffic signs are very important where it has the highest bar, as well as the traffic lights that the bar of âvery importantâ also has the majority of the answers and the last bar graph which shows the importance of traffic enforcers, the votes too close to each other but the âvery importantâ has still the highest bar.

```python
importance_of_signs_count = research_data_path[['How important is traffic signs in your area:']].value_counts()
importance_of_lights_count = research_data_path[['How important of traffic lights in your area:']].value_counts()
importance_of_enforcers_count = research_data_path[['How important of traffic enforcers in your area:']].value_counts()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
importance_of_signs_count.plot(kind='bar',ax=axes[0],figsize=(20,5))
importance_of_lights_count.plot(kind='bar',ax=axes[1],figsize=(20,5))
importance_of_enforcers_count.plot(kind='bar',ax=axes[2],figsize=(20,5))
```
