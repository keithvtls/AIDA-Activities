```python
#TASK 2, PART 2: 
#MEAN, MEDIAN, MODE, STANDARD DEVIATION

import numpy as np

def get_class_stats(overall_grades):
  prelim_mean = overall_grades['prelim_grades'].mean(axis=0)
  prelim_median = overall_grades['prelim_grades'].median(axis=0)
  prelim_mode = overall_grades['prelim_grades'].mode(dropna=True)
  prelim_std = overall_grades['prelim_grades'].std(ddof=0)

  midterm_mean = overall_grades['midterm_grades'].mean(axis=0)
  midterm_median = overall_grades['midterm_grades'].median(axis=0)
  midterm_mode = overall_grades['midterm_grades'].mode(dropna=True)
  midterm_std = overall_grades['midterm_grades'].std(ddof=0)

  final_mean = overall_grades['final_grades'].mean(axis=0)
  final_median = overall_grades['final_grades'].median(axis=0)
  final_mode = overall_grades['final_grades'].mode(dropna=True)
  final_std = overall_grades['final_grades'].std(ddof=0)

  sem_mean = overall_grades['semestral_grades'].mean(axis=0)
  sem_median = overall_grades['semestral_grades'].median(axis=0)
  sem_mode = overall_grades['semestral_grades'].mode(dropna=True)
  sem_std = overall_grades['semestral_grades'].std(ddof=0)
  #sem_std = 1

  df_stats = pd.DataFrame({'Statistics': ['mean', 'median', 'mode', 'standard deviation'],
                           'Prelim': [prelim_mean, prelim_median, prelim_mode, prelim_std],
                           'Midterm': [midterm_mean, midterm_median, midterm_mode, midterm_std],
                           'Finals': [final_mean, final_median, final_mode, final_std],
                           'Semestral': [sem_mean, sem_median, sem_mode, sem_std]})

  return(df_stats)
```

```python
df_overall_stats = get_class_stats(df_class_grades)
df_overall_stats
```

** **Note:** due to data privacy, the data privacy office has mandated your code to refrain from printing the student number and names of students. Their student numbers and names  must be masked with asterisks.

```python
def get_grades(p_grade, m_grade, f_grade):
    return (p_grade * 0.30) + (m_grade * 0.30) + (f_grade * 0.40)
    
df_class_grades['semestral_grades'] = get_grades(df_class_grades['prelim_grades'], df_class_grades['midterm_grades'], df_class_grades['final_grades'])
```

```python
class Section:

    def __init__(self,term):
        self.term_grade = term
        
    def get_failed(self):
        self.class_grades = df_class_grades
        self.check_fail = self.class_grades[(self.class_grades[self.term_grade] < 70)]
        self.show_students = self.check_fail[['name', self.term_grade]]
        return self.show_students
    
    def fail_count(self):
        return len(self.check_fail)
```

```python
term = 'prelim_grades' # By changing the term, the output will reveal who failed each term
failed_students = Section(term)
failed_students.get_failed()
```

```python
print(f'Number of students who failed in {term} is {failed_students.fail_count()}')
```
